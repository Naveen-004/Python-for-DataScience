```python
import pandas as pd
import numpy as np

df = pd.read_csv('DataFrames/weather.csv')
df
```

```python
print(df)
df1 = df.dropna()
df1
```

```python
df2 = df.dropna(how='any')
df2
```

```python
print(df)
df3 = df.dropna(how='all')
df3
```

```python
df4 = df.dropna(how='all', thresh= 3)
df4
```

```python
df5 = df.fillna(method='bfill')
df5
```

```python
df6 = df.fillna(method='ffill', limit = 1)
df6
```

```python
mean = df.fillna(df.mean())
mean
```

```python
median = df.fillna(df.median())
median
```

```python
mode=df.fillna(df.mode())
mode=df['event'].fillna(df['event'].mode()[0])
mode
```
